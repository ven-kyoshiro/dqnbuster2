```python
import numpy as np
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
```

```python
f_name = 'recorded1.npy'
data = np.load(f_name).T
data.shape
```

```python
plt.plot(data[:,0])
plt.plot(data[:,1])
```

```python
normd_data = ((data-np.mean(data))/np.std(data)).clip(-4,4)
```

```python
scaled = np.int16(normd_data/np.max(np.abs(normd_data)) * 32767)
```

```python
write(f_name.replace('.npy','.wav'), 4000, scaled)
```
